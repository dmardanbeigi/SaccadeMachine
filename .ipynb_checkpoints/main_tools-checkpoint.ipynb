# Smaller functions


```python

def AngleBetweenTwoVectors(vec1,vec2):
    angle=np.rad2deg( np.arccos(np.dot(vec1,vec2)/(np.linalg.norm(vec1)*np.linalg.norm(vec2))))
    return angle

def pixels_to_degrees_center_based(x,distance_to_screen=None,resolution=None,monitor_size=None):
    # amp=atan(Z/f)
    # where f is the distance to the screen and Z is defined as:
    # deltaX*monitor_width_in_cm / monitor_width_in_pixels

    if type(distance_to_screen)==type(None):
        distance_to_screen=EXPERIMENT_SETTINGS['screen_distance']
    if type(resolution)==type(None):
        resolution=EXPERIMENT_SETTINGS['screen_res']
    if type(monitor_size)==type(None):
        monitor_size=EXPERIMENT_SETTINGS['screen_size']
    
    if type(x)!=list:
        out=np.degrees(np.arctan2(x * monitor_size[0]/resolution[0] , distance_to_screen) ) 
    else:
        out=np.degrees(np.arctan2( np.array(x) * monitor_size[0]/resolution[0] , np.array([distance_to_screen]*len(np.array(x)) ) ))

    return out

def pixels_to_degrees_precise(dx=None,dy=None,resxy_av=None):
    
#     print(dx,type(dx))
#     print(resxy_av)
    
    if type(dx)!=list:
        out=math.hypot(dx/resxy_av[0], dy/resxy_av[1]) 
    else:
        out=[]
        for x_i,x in dx:   
            out=math.hypot(x/resxy_av[x_i][0], dy[x_i]/resxy_av[x_i][1])
        out=np.array(out)
        
    
    return out


def rotatePoint(centerPoint,point,angle):
    """Rotates a point around another centerPoint. Angle is in degrees.
    Rotation is counter-clockwise"""
    angle = math.radians(angle)
    temp_point = point[0]-centerPoint[0] , point[1]-centerPoint[1]
    temp_point = ( temp_point[0]*math.cos(angle)-temp_point[1]*math.sin(angle) , temp_point[0]*math.sin(angle)+temp_point[1]*math.cos(angle))
    temp_point = temp_point[0]+centerPoint[0] , temp_point[1]+centerPoint[1]
    return temp_point

def CalculateAngle2(p1,p2):
    angle=math.degrees(math.atan2(p2[1]-p1[1], p2[0]-p1[0]))
    if angle<0:
        angle=-angle
    else:
        angle=(360-angle)
    return angle

# in_radius
def in_radius(c, r, x):
    try:
        return int(math.hypot(c[0]-x[0], c[1]-x[1]) <= r )
    except:
        return 0

# compare_two_lists
compare_two_lists = lambda a,b: len(a)==len(b) and len(a)==sum([1 for i,j in zip(a,b) if i==j])
```

# Load & save functions

```python
# Load & save table

def LoadTable(file,path=''):
    if path=='':
        path=EXPERIMENT_FOLDER_PATH
        
    file=file+'.sm'
    df=None
    fileList= [os.path.join(path, f) for f in os.listdir(path) if os.path.isfile(os.path.join(path, f)) and (file in f) ]
    fileList=sorted(fileList, key=str.lower)
    for f in fileList:
        df=pd.read_pickle(f) if df is None else pd.concat([df,pd.read_pickle(f)], ignore_index=True)
        df=df.reset_index(drop=True)
    return df
    
    
    

def SaveTable(df,file,path=''):
    if path=='':
        path=EXPERIMENT_FOLDER_PATH
        
        
    file=file+'.sm'
    ## Remove all the pkl files from before
    fileList= [os.path.join(path, f) for f in os.listdir(path) if os.path.isfile(os.path.join(path, f)) and (file in f) ]
    for f in fileList:
        os.remove(f)
    
    ##split table and save
#     print('saving table with %s rows (check if it loads the same amount)'%df.shape[0])
    chunk_size = 1000
    start = 0
    end = chunk_size-1
    count=0
    if end>=df.shape[0]:
        df.to_pickle('%s%s.%s' %(path,file,count) )
    else:
        while end < df.shape[0]: 
            count=count+1
            chunk = df.iloc[start:end]
            chunk.to_pickle('%s%s.%s' %(path,file,str(count)) )
#             display(chunk[chunk.name=='cont_jrackie'])
     
            start += chunk_size-1
            end += chunk_size-1
        if ((df.shape[0]-1)>start):
            count=count+1
            chunk = df.iloc[start+1:df.shape[0]]
            chunk.to_pickle('%s%s.%s' %(path,file,str(count)) )
#             display(chunk[chunk.name=='cont_jrackie'])


```

# Filter

```python
def Filter(fixation_saccade_table,gaze_quality_table):

    selected_blocks=[]
    selected_groups=[]
    
    fixation_saccade_table['EXCLUDE']=False  
    gaze_quality_table['EXCLUDE']=False  
    fixation_saccade_table['EXCLUDE_NOTE']=''  
    gaze_quality_table['EXCLUDE_NOTE']=''  

    
    for g in group_checkbox:
        if g.value:
            selected_groups.append(g.description.split('(')[0][:-1])

    for g in block_checkbox:
        if g.value:
            selected_blocks.append(float(g.description.split('(')[0][6:-1]))

    GF=~fixation_saccade_table['SUBJECT_GROUP'].isin(selected_groups)
    BF=~fixation_saccade_table['BLOCK'].isin(selected_blocks)
    fixation_saccade_table.loc[(GF)|(BF),'EXCLUDE']=True
    fixation_saccade_table.loc[(GF)|(BF),'EXCLUDE_NOTE']='participant group or block filtered'


    
#     display(fixation_saccade_table.loc[~fixation_saccade_table['BLOCK'].isin(selected_blocks),:])
    gaze_quality_table.loc[~gaze_quality_table['SUBJECT_GROUP'].isin(selected_groups),'EXCLUDE']=True
    gaze_quality_table.loc[~gaze_quality_table['SUBJECT_GROUP'].isin(selected_groups),'EXCLUDE_NOTE']='participant group filtered'


    
    ## FILTER SMALL SACCADES (PERHAPS JUST ADD THEM TO THE EXCLUDE COLUMN)
    
    global MinSacAmp
    fixation_saccade_table.loc[fixation_saccade_table['SACCADE_AMPLITUDE']<MinSacAmp,'EXCLUDE']=True
    fixation_saccade_table.loc[fixation_saccade_table['SACCADE_AMPLITUDE']<MinSacAmp,'EXCLUDE_NOTE']='saccade amplitude smaller than %s deg threshold'%MinSacAmp
    
    
    ## exclude trials that have any of the msgs missing
    msg_missing=pd.isnull(fixation_saccade_table.loc[:,['FIXATION_TARGET_ONSET_INDEX','TARGET_ONSET_INDEX',
         'FIXATION_TARGET_OFFSET_INDEX','TARGET_OFFSET_INDEX']]).any(axis=1)
    fixation_saccade_table.loc[msg_missing,'EXCLUDE']=True
    fixation_saccade_table.loc[msg_missing,'EXCLUDE_NOTE']='SAMPLE_MESSAGEs missing in the recording'
    

    return fixation_saccade_table,gaze_quality_table
    
```

```python
# Process
def Process(path):
    '''
    - process all files if hasn't been done before
    - remove subjects from the existing tables if the original file for the subject is removed from the dataset folder
    - only process new subjects that are added to the dataset folder and add them to the existing table
    '''
    global FIXATIONSACCADE_TABLE,GAZE_QUALITY_TABLE
    
    
    ## create folder for the experiment 
    if not os.path.exists(EXPERIMENT_FOLDER_PATH):
        with out_load_details:
            print('processed files will be stored in %s'% EXPERIMENT_FOLDER_PATH)
        os.makedirs(EXPERIMENT_FOLDER_PATH)
    
    ## SAVE SETTING (this is only effective on first time)
    SaveSettings()
    Display_Settings()
    
    ## Load already processed tables
    FIXATIONSACCADE_TABLE=LoadTable('FIXATIONSACCADE_TABLE')
    GAZE_QUALITY_TABLE=LoadTable('GAZE_QUALITY_TABLE')
   
    Existing_FIXATIONSACCADE_TABLE=FIXATIONSACCADE_TABLE
    Existing_GAZE_QUALITY_TABLE=GAZE_QUALITY_TABLE
   

    
    ## Get a list of all participants inside the dataset folder
    fileList= np.array([(f[:-4],os.path.join(path, f)) for f in os.listdir(path) if os.path.isfile(os.path.join(path, f)) and ((f[-3:]=='csv') or (f[-4:]=='xlsx')) ])
    
    
    try:
        fileList=SUBSET_FOR_ETRA(path)
    except:
        pass

    
    if (len(fileList)==0):
        return False

    
    N_removed=0
    N_new=0
    tobe_processed_names=[] 
    subjects_with_error=[]
    
    ## check if any processed table already exists
    if Existing_GAZE_QUALITY_TABLE is not None:
        ## load the table
#         print('Looking at the dataset folder for any differences...')
        N_before=len(Existing_GAZE_QUALITY_TABLE['SUBJECT_NAME'].unique())
    
    
        ## remove subject from the table if it's been removed from the folder
        Existing_FIXATIONSACCADE_TABLE=Existing_FIXATIONSACCADE_TABLE[Existing_FIXATIONSACCADE_TABLE['SUBJECT_NAME'].isin(fileList[:,0])]
        Existing_GAZE_QUALITY_TABLE=Existing_GAZE_QUALITY_TABLE[Existing_GAZE_QUALITY_TABLE['SUBJECT_NAME'].isin(fileList[:,0])]
        N_removed=N_before-len(Existing_GAZE_QUALITY_TABLE['SUBJECT_NAME'].unique())

            
        ## participants that exist in both the table and the folder
        already_processed_subjects=list(filter(lambda itm:itm in Existing_GAZE_QUALITY_TABLE['SUBJECT_NAME'].unique(),fileList[:,0]))
        ## remove them from the fileList so that we don't process them again (we could reprocess all again but we can save time this way)
        fileList2=list(filter(lambda x:x[0] not in already_processed_subjects,list(fileList)))



 
        ## only extract data from the new subjects found in the folder
        FixationSaccade_table,gaze_quality_table=Generate_Tables(fileList2)
        if 'SUBJECT_NAME' in gaze_quality_table.columns:
            subjects_with_error=[x[0] for x in fileList2 if x[0] not in gaze_quality_table['SUBJECT_NAME'].unique()]

            
        FixationSaccade_table=pd.concat([Existing_FIXATIONSACCADE_TABLE,FixationSaccade_table], ignore_index=True) 
        gaze_quality_table=pd.concat([Existing_GAZE_QUALITY_TABLE,gaze_quality_table], ignore_index=True) 
        

        if N_removed>0:
            with out_load_details:
                print('\n %s subjects removed from the dataset'%(N_removed))
        
        N_new=len(fileList2)-len(subjects_with_error)


            
    else:
        ## extract and process all participants
        
        FixationSaccade_table,gaze_quality_table=Generate_Tables(fileList)
        _lst=gaze_quality_table['SUBJECT_NAME'].unique() if 'SUBJECT_NAME' in gaze_quality_table.columns.values else []
        subjects_with_error=[x[0] for x in fileList if x[0] not in _lst]
    
        N_new=len(fileList)-len(subjects_with_error)
    
    
    if (N_new>0) or (N_removed>0):
        FIXATIONSACCADE_TABLE=FixationSaccade_table
        GAZE_QUALITY_TABLE=gaze_quality_table


        

        SaveTable(FIXATIONSACCADE_TABLE,'FIXATIONSACCADE_TABLE')
        SaveTable(GAZE_QUALITY_TABLE,'GAZE_QUALITY_TABLE')


    if N_new>0:
        with out_load_details:
            print('\n %s new subjects found in the dataset. '%(N_new))
            
            
#         print([x for x in tobe_processed_names if not x in subjects_with_error])
    if len(subjects_with_error)>0:
        with out_load_details:
            print('\n The following subjects cannot be processed: \n',subjects_with_error)
    
    
    
    if type(GAZE_QUALITY_TABLE)==type(None):
        return False
    
    if len(GAZE_QUALITY_TABLE)==0:
        return False
    


    
    return True
```

## SetTargetAppearanceCondition

```python

def SetTargetAppearanceCondition(data):
    
    data['CONDITION_APPEARANCE']=''

    df=data.groupby(['TRIAL_INDEX','SECOND_DISPLAY']).first().loc[:,[col for col in data.columns.values[ data.columns.str.contains('^TARGET_X|^TARGET_Y|^DISTRACTOR_.*_X|^DISTRACTOR_.*_Y')]]].reset_index()

    
    ## number of distractors
    N_distractors=df.columns.str.contains('^DISTRACTOR_.*_X').sum()
    N_targets=df.columns.str.contains('^TARGET_X').sum()

    cmp=lambda a,b:(a>b)-(a<b)
    for index, row in df.iterrows():
        
        ## first trial
        if row['TRIAL_INDEX']==1 and row['SECOND_DISPLAY']==0:
            if N_targets==1:
                data.loc[data['TRIAL_INDEX']==row['TRIAL_INDEX'],'CONDITION_APPEARANCE']='TN'## default
                continue
            if N_targets==0 and N_distractors==1:# normal antisaccade 
                data.loc[data['TRIAL_INDEX']==row['TRIAL_INDEX'],'CONDITION_APPEARANCE']='DN'## defaultabs
                continue
                
        ## other trials
        if N_targets==1:

            data.loc[data['TRIAL_INDEX']==row['TRIAL_INDEX'],'CONDITION_APPEARANCE']='TN'## default
            if cmp( (float(row['TARGET_X']),float(row['TARGET_Y'])),
                   (float(df.shift().loc[index,'TARGET_X']),float(df.shift().loc[index,'TARGET_Y']) ))==0:
                data.loc[data['TRIAL_INDEX']==row['TRIAL_INDEX'],'CONDITION_APPEARANCE']='TT'
            else:
                for dis in range(1,N_distractors+1):
                    if cmp( (float(row['TARGET_X']),float(row['TARGET_Y'])),
                           (float(df.shift().loc[index,'DISTRACTOR_%s_X'%dis]),float(df.shift().loc[index,'DISTRACTOR_%s_Y'%dis]) ))==0:
                        data.loc[data['TRIAL_INDEX']==row['TRIAL_INDEX'],'CONDITION_APPEARANCE']='TD%s'%dis         
                        break

            continue

        if N_targets==0 and N_distractors==1:# normal antisaccade 
            data.loc[data['TRIAL_INDEX']==row['TRIAL_INDEX'],'CONDITION_APPEARANCE']='DN'## default
            if cmp( (float(row['DISTRACTOR_1_X']),float(row['DISTRACTOR_1_Y'])),
                   (float(df.shift().loc[index,'DISTRACTOR_1_X']),float(df.shift().loc[index,'DISTRACTOR_1_Y']) ))==0:
                data.loc[data['TRIAL_INDEX']==row['TRIAL_INDEX'],'CONDITION_APPEARANCE']='DD'
            continue
    return data
```

## Saccade and fixations

```python
def ProcessFixation(df,trial_index,trial_is_second_display):
    first_row=df.iloc[0]

    _row_dic={}
    
    _row_dic.update({'BLOCK':first_row['BLOCK']})
    _row_dic.update({'CONDITION':first_row['CONDITION']})
    _row_dic.update({'CONDITION_APPEARANCE':first_row['CONDITION_APPEARANCE']})
    _row_dic.update({'TRIAL_INDEX':trial_index})
    _row_dic.update({'SECOND_DISPLAY':trial_is_second_display})
    
    df_tmp=_tmp_[(_tmp_keys_[:,0]==first_row['BLOCK']) & (_tmp_keys_[:,1]==trial_index) & (_tmp_keys_[:,2]==trial_is_second_display)]
    if len(df_tmp)>0:
        for col in list(df_tmp.columns.values[df_tmp.columns.str.contains('^TARGET|^DISTRACTOR|^FIXATION')]):
            
            _row_dic.update({col:df_tmp.iloc[0][col]})


    
    _row_dic.update({'EVENT':'fixation'})
    if 'GAZE_X' in df.columns.values:
        pos=(df['GAZE_X'].mean(),df['GAZE_Y'].mean())
    else:
        pos=[None,None]
    
    _row_dic.update({'FIXATION_MEAN_X':pos[0]})
    _row_dic.update({'FIXATION_MEAN_Y':pos[1]})

    _row_dic.update({'FIXATION_START_INDEX':df.index.values[0]})
    _row_dic.update({'FIXATION_END_INDEX':df.index.values[-1]})

    _row_dic.update({'FIXATION_START_TIMESTAMP':df['TIMESTAMP'].values[0]})
    _row_dic.update({'FIXATION_END_TIMESTAMP':df['TIMESTAMP'].values[-1]})

    

            
    r=EXPERIMENT_SETTINGS['AOI_size']/2

    trial_pre=GetRowFromPreviousSubTrial(trial_index,first_row['BLOCK'],trial_is_second_display)
    
    ## relation to center  
    _row_dic.update({'FIXATION_ON_CENTER':in_radius(SCREEN_CENTER, r, pos)})
    _row_dic.update({'FIXATION_DISTANCE_TO_CENTER':np.linalg.norm([pos[0]-SCREEN_CENTER[0],pos[1]-SCREEN_CENTER[1]])})


    
    ## relation to target    
    if 'TARGET_X' in df.columns.values:
        _row_dic.update({'FIXATION_ON_TARGET':in_radius((first_row['TARGET_X'],first_row['TARGET_Y']), r, pos)})
        _row_dic.update({'FIXATION_DISTANCE_FROM_TARGET':distance.euclidean(pos,(first_row['TARGET_X'],first_row['TARGET_Y']))})

        ## how about previous trial
        if len(trial_pre)==0:
            _row_dic.update({'FIXATION_ON_PREVIOUS_TARGET':0})

        else:
            _row_dic.update({'FIXATION_ON_PREVIOUS_TARGET':in_radius((trial_pre['TARGET_X'],trial_pre['TARGET_Y']), r, pos)})

    else:
        _row_dic.update({'FIXATION_ON_TARGET':0})
        _row_dic.update({'FIXATION_ON_PREVIOUS_TARGET':0})
        _row_dic.update({'FIXATION_DISTANCE_FROM_TARGET':np.nan})


    ## relation to distractors
    for col in df.columns.values:
        if col.startswith('DISTRACTOR_') and col.endswith('_X'):

            _row_dic.update({'FIXATION_ON_%s'%col[:-2]:in_radius((first_row[col],first_row['%s_Y'%col[:-2]]), r, pos)})
            ## how about previous trial
            if len(trial_pre)==0:
                _row_dic.update({'FIXATION_ON_PREVIOUS_%s'%col[:-2]:0})

            else:
                _row_dic.update({'FIXATION_ON_PREVIOUS_%s'%col[:-2]:in_radius((trial_pre[col],trial_pre['%s_Y'%col[:-2]]), r, pos)})


    return _row_dic
    
def ProcessSaccade(df,trial_index,trial_is_second_display):
    first_row=df.iloc[0]

    _row_dic={}
    _row_dic.update({'BLOCK':first_row['BLOCK']})
    _row_dic.update({'CONDITION':first_row['CONDITION']})
    _row_dic.update({'CONDITION_APPEARANCE':first_row['CONDITION_APPEARANCE']})
    _row_dic.update({'TRIAL_INDEX':trial_index})
    _row_dic.update({'SECOND_DISPLAY':trial_is_second_display})
    
    df_tmp=_tmp_[(_tmp_keys_[:,0]==first_row['BLOCK']) & (_tmp_keys_[:,1]==trial_index) & (_tmp_keys_[:,2]==trial_is_second_display)]
    if len(df_tmp)>0:
        for col in list(df_tmp.columns.values[df_tmp.columns.str.contains('^TARGET|^DISTRACTOR|^FIXATION')]):
            _row_dic.update({col:df_tmp.iloc[0][col]})



            
    A=[df['GAZE_X'].values[0],df['GAZE_Y'].values[0]]
    B=[df['GAZE_X'].values[-1],df['GAZE_Y'].values[-1]]
    
    _row_dic.update({'EVENT':'saccade'})
    
    _row_dic.update({'SACCADE_A_X':A[0]})

    _row_dic.update({'SACCADE_A_Y':A[1]})

    _row_dic.update({'SACCADE_B_X':B[0]})

    _row_dic.update({'SACCADE_B_Y':B[1]})


    amp_pix=distance.euclidean(B,A)
    amp_deg=pixels_to_degrees_center_based(amp_pix)
    
    if 0:## use this only if RESOLUTION data are correct. Usually RESOLUTION_X values people get from eyelink are not accurate as people forget to set the phisical.ini file properly before their recordings!
        if 'RESOLUTION_X' in  df.columns.values:
            resxy_av=(df['RESOLUTION_X'].mean(),df['RESOLUTION_Y'].mean())
            amp_deg=pixels_to_degrees_precise( dx=B[0]-A[0],
                                             dy=B[1]-A[1],
                                             resxy_av=resxy_av) 

    _row_dic.update({'SACCADE_AMPLITUDE':amp_deg})

    _row_dic.update({'SACCADE_AMPLITUDE_PIX':amp_pix})

    _row_dic.update({'SACCADE_ANGLE':CalculateAngle2(A,B)})
    _row_dic.update({'SACCADE_START_INDEX':df.index.values[0]})
    _row_dic.update({'SACCADE_END_INDEX':df.index.values[-1]})
    _row_dic.update({'SACCADE_START_TIMESTAMP':df['TIMESTAMP'].values[0]})
    _row_dic.update({'SACCADE_END_TIMESTAMP':df['TIMESTAMP'].values[-1]})

    trial_pre=GetRowFromPreviousSubTrial(trial_index,first_row['BLOCK'],trial_is_second_display)


    _row_dic=ProcessSaccade_part2(_row_dic,trial_pre,first_row,A,B)
            
    return _row_dic



def Fix2Fix_Saccade(first_row,row,F1,F2):
         
    _row_dic={}

    A=[F1['FIXATION_MEAN_X'],F1['FIXATION_MEAN_Y']]
    B=[F2['FIXATION_MEAN_X'],F2['FIXATION_MEAN_Y']]
    
    
    _row_dic.update({'EVENT':'saccade'})
    
    _row_dic.update({'SACCADE_A_X':A[0]})

    _row_dic.update({'SACCADE_A_Y':A[1]})

    _row_dic.update({'SACCADE_B_X':B[0]})

    _row_dic.update({'SACCADE_B_Y':B[1]})


    amp_pix=distance.euclidean(B,A)
    amp_deg=pixels_to_degrees_center_based(amp_pix)
    
    _row_dic.update({'SACCADE_AMPLITUDE':amp_deg})

    _row_dic.update({'SACCADE_AMPLITUDE_PIX':amp_pix})

    _row_dic.update({'SACCADE_ANGLE':CalculateAngle2(A,B)})
    _row_dic.update({'SACCADE_START_INDEX':F1['FIXATION_END_INDEX']})
    _row_dic.update({'SACCADE_END_INDEX':F1['FIXATION_START_INDEX']})
    _row_dic.update({'SACCADE_START_TIMESTAMP':F1['FIXATION_END_TIMESTAMP']})
    _row_dic.update({'SACCADE_END_TIMESTAMP':F1['FIXATION_START_TIMESTAMP']})

    trial_pre=GetRowFromPreviousSubTrial(F1['TRIAL_INDEX'],F1['BLOCK'],F1['SECOND_DISPLAY'])

    
    
    
    _row_dic=ProcessSaccade_part2(_row_dic,trial_pre,first_row,A,B)
    
    for k,v in _row_dic.items():
        row.update({k:v})
    

    return row

def ProcessSaccade_part2(_row_dic,trial_pre,first_row,A,B):
        ## all targets in the screen 
    
    targets_to_check=[]
    targets_to_check_pre=[]
    

    _row_dic.update({'SACCADE_TOWARDS_CENTER':0})

    
    if 'TARGET_X' in first_row.keys():
        _row_dic.update({'SACCADE_TOWARDS_TARGET':0})
        targets_to_check.append([first_row['TARGET_X'],first_row['TARGET_Y']])
 
        _row_dic.update({'SACCADE_END_DISTANCE_FROM_TARGET':distance.euclidean(B,(first_row['TARGET_X'],first_row['TARGET_Y']))})

        _row_dic.update({'SACCADE_TOWARDS_PREVIOUS_TARGET':0})
        if len(trial_pre)!=0:
            targets_to_check_pre.append([trial_pre['TARGET_X'],trial_pre['TARGET_Y']])

    else:
        _row_dic.update({'SACCADE_TOWARDS_TARGET':0})
        _row_dic.update({'SACCADE_TOWARDS_PREVIOUS_TARGET':0})
        _row_dic.update({'SACCADE_END_DISTANCE_FROM_TARGET':0})


    distractors_to_check=[]
    distractors_to_check_pre=[]
    opposites_to_check=[]
    opposites_to_check_pre=[]
    
    def GetOpposite(D):
        oppos=rotatePoint(SCREEN_CENTER,D,180)
        ## shift the point to a far distance.
        scale=10
        oppos=np.array([(oppos[0]-SCREEN_CENTER[0])*scale,(oppos[1]-SCREEN_CENTER[1])*scale])+SCREEN_CENTER

        return oppos
    
    for col in first_row.keys():
        if col.startswith('DISTRACTOR_') and col.endswith('_X'):
            
            distractor_i=col.split('_')[1]
            
            _row_dic.update({'SACCADE_TOWARDS_%s'%col[:-2]:0})
            distractors_to_check.append([first_row[col],first_row['%s_Y'%col[:-2]]])
            
 
            if EXPERIMENT_SETTINGS['type']=='Anti-Saccade':
                _row_dic.update({'SACCADE_TOWARDS_OPPOSITE_%s'%distractor_i:0})
                opposites_to_check.append(GetOpposite([first_row[col],first_row['%s_Y'%col[:-2]]]))
                

            ## how about previous trial
            if len(trial_pre)==0:
                _row_dic.update({'SACCADE_TOWARDS_PREVIOUS_%s'%col[:-2]:0})
      
                if EXPERIMENT_SETTINGS['type']=='Anti-Saccade': 
                    _row_dic.update({'SACCADE_TOWARDS_PREVIOUS_OPPOSITE_%s'%distractor_i:0})


            else:
                _row_dic.update({'SACCADE_TOWARDS_PREVIOUS_%s'%col[:-2]:0})
                distractors_to_check_pre.append([trial_pre[col],trial_pre['%s_Y'%col[:-2]]])
                
                if EXPERIMENT_SETTINGS['type']=='Anti-Saccade':
                    _row_dic.update({'SACCADE_TOWARDS_PREVIOUS_OPPOSITE_%s'%distractor_i:0})
                    opposites_to_check_pre.append(GetOpposite([trial_pre[col],trial_pre['%s_Y'%col[:-2]]]))
                
                    
                


    TOI=GetSaccadeTOI(A,B,[SCREEN_CENTER]+targets_to_check+distractors_to_check+opposites_to_check)
    TOI_pre=GetSaccadeTOI(A,B,[SCREEN_CENTER]+targets_to_check_pre+distractors_to_check_pre+opposites_to_check_pre)

    
    def PorcessIndex(_row_dic,i,pre=''):
        if not np.isnan(i):
            if i==0 and pre=='': ## TOI was center
                _row_dic.update({'SACCADE_TOWARDS_CENTER':1})
            elif len(targets_to_check)==1 and i==1: ## TOI was target
                _row_dic.update({'SACCADE_TOWARDS%s_TARGET'%pre:1})
            elif len(distractors_to_check)>0 and i>=(1+len(targets_to_check)) and i<(1+len(targets_to_check)+len(distractors_to_check)):## TOI was one of the distractors
                _row_dic.update({'SACCADE_TOWARDS%s_DISTRACTOR_%s'%(pre,i-(1+len(targets_to_check))+1):1})
            elif len(opposites_to_check)>0 and i>=(1+len(targets_to_check)+len(distractors_to_check)) and i<(1+len(targets_to_check)+len(distractors_to_check)+len(opposites_to_check)):## TOI was one of the opposites
                _row_dic.update({'SACCADE_TOWARDS%s_OPPOSITE_%s'%(pre,i-(1+len(targets_to_check)+len(distractors_to_check))+1):1})
        return _row_dic

    _row_dic=PorcessIndex(_row_dic,TOI,pre='')
    _row_dic=PorcessIndex(_row_dic,TOI_pre,pre='_PREVIOUS')
    


    return _row_dic

def GetSaccadeTOI(A,B,targets):
    A=np.array(A)
    B=np.array(B)
    targets=np.array(targets)

    rat=2/3
    rat2=0.5
    ## I ended up using the following values. when beta and alpha are equal C4 is not functional anymore
    alpha=30# 10
    beta=30#20 
    
    
    
    ##find point M
    M=np.array([(B[0]-A[0])*rat,(B[1]-A[1])*rat])+A
    
    ## find another example point on the line perpendicular to saccade from M
    slope_sac=(B[1]-A[1])/(B[0]-A[0])
    if slope_sac!=0:
        slope_perpend=-1/slope_sac
        M2=np.array([0,slope_perpend*(0-M[0])+M[1]])
    else:
        M2=np.array([M[0],0])
        
    dists_to_Mline_B=np.linalg. norm(np.cross(M2-M, M-B))/np.linalg. norm(M2-M)
    
    C1=[]
    C2=[]
    dists_to_B=[]
    dists_to_Mline=[]
    
    C4=[]
    
    for T in targets:
        ## C1
        if slope_sac!=0:
            slope_perpend=-1/slope_sac
            T_rel_perpend=T[1]-slope_perpend*(T[0]-M[0])-M[1]
            B_rel_perpend=B[1]-slope_perpend*(B[0]-M[0])-M[1]
        else:
            T_rel_perpend=T[1]-M[0]
            B_rel_perpend=B[1]-M[0]

            
        c2=T_rel_perpend*B_rel_perpend
#         print(c2)
        C1.append(T_rel_perpend*B_rel_perpend>0)

        dists_to_Mline.append( np.linalg. norm(np.cross(M2-M, M-T))/np.linalg. norm(M2-M))


        ## C2
        C2.append(AngleBetweenTwoVectors(T-A,B-A)<alpha)
#         print(AngleBetweenTwoVectors(T-A,B-A))
        ## C3
        if C1[-1]*C2[-1]:
            dists_to_B.append(distance.euclidean(T,B))
        else:
            dists_to_B.append(np.nan)
    
        ## C4

        C4.append(AngleBetweenTwoVectors(T-A,B-A)<beta if C2[-1]==False else False)
        
        
        
    C1=np.array(C1)
    C2=np.array(C2)
    C4=np.array(C4)

    dists_to_B=np.array(dists_to_B)
    dists_to_Mline=np.array(dists_to_Mline)

    ## candidate
    if (np.isnan(dists_to_B)).all() :
        candidate_index=np.nan
    else:
        candidate_index=np.nanargmin(dists_to_B)
    

#     print('C1:',C1)
#     print('C2:',C2)
#     print('C4:',C4)
#     print('dists_to_B:',dists_to_B)
#     print('dists_to_Mline:',dists_to_Mline)
#     print('candidate_index:',candidate_index)
    
    ## Check the distance of all targets that are true in C4
    if not np.isnan( candidate_index):
        for T_ind,_c4 in enumerate(C4):
            if _c4: 
                if (dists_to_Mline[T_ind]<dists_to_Mline[candidate_index]) and (dists_to_Mline[T_ind]>dists_to_Mline_B*rat2):
                    candidate_index=np.nan


    
#     print('candidate_index:',candidate_index)
    return candidate_index
```

# GazeQuality

```python
# CleanGazeData

def CleanGazeData(table):

    table_gaze_quality= pd.DataFrame( )
    
    table_gaze_quality.at[0,'TOTAL_ROWS'] = len( table)

    
    df=table.copy(deep=True)
    df['FILTERED']=0


    ## remove where gaze=(0,0) (already filtered in eyelink data)
    whereIsZero=(df['GAZE_X']==0) | (df['GAZE_Y']==0)
    df.loc[whereIsZero,'FILTERED']=1
    table_gaze_quality.at[0,'ROWS_WITH_ZERO_GAZE'] = sum( whereIsZero )


    
    ## remove blink data (if column exists)
    if 'IN_BLINK' in df.columns.values:
        whereIsBlink=df['IN_BLINK']==1
        df.loc[whereIsBlink,'FILTERED']=1
        table_gaze_quality.at[0,'ROWS_IN_BLINK'] =sum( whereIsBlink )


        
    return df,table_gaze_quality

def DoubleCheckGazeQuality(FixationSaccade_table,gaze_quality_table):
    ## check distance from center before TARGET_ONSET
    
    distances=[]
    for (a,b,c),d in FixationSaccade_table.groupby(['TRIAL_INDEX','SECOND_DISPLAY','BLOCK']):
        distances.append(d[d['FIXATION_START_TIMESTAMP']<d['FIXATION_TARGET_OFFSET_TIMESTAMP']]['FIXATION_DISTANCE_TO_CENTER'].median())

    gaze_quality_table.at[0,'AVERAGE_FIXATION_DISTANCE_TO_CENTER']= np.nanmean(distances)

    ## scores
    gaze_quality_table['SCORE1']=1- gaze_quality_table['ROWS_WITH_ZERO_GAZE'] /gaze_quality_table['TOTAL_ROWS'] 
    if 'ROWS_IN_BLINK' in gaze_quality_table.columns.values:
        gaze_quality_table['SCORE2']=1- 1.3*(gaze_quality_table['ROWS_IN_BLINK'] /gaze_quality_table['TOTAL_ROWS'] )
    else:
        gaze_quality_table['SCORE2']=1
    maximum_acceptable_dist_to_center=EXPERIMENT_SETTINGS['AOI_size']/2# pixels
    gaze_quality_table['SCORE3']=1-gaze_quality_table['AVERAGE_FIXATION_DISTANCE_TO_CENTER']/maximum_acceptable_dist_to_center
    gaze_quality_table['SCORE_AV']=(gaze_quality_table['SCORE1']+gaze_quality_table['SCORE2']+gaze_quality_table['SCORE3'])/3

    gaze_quality_table['SCORE_AV']=np.clip(gaze_quality_table['SCORE_AV'].values, 0.1, 1)
    
    return gaze_quality_table
```

```python
# PlotGazeQuality
def PlotGazeQuality():
    mpld3.enable_notebook()


    if GAZE_QUALITY_TABLE is not None:
        
        tmp=GAZE_QUALITY_TABLE.loc[(GAZE_QUALITY_TABLE['EXCLUDE']==0) , ['SUBJECT_NAME','SCORE_AV', 'SUBJECT_GROUP'] ]

        
        tmp=tmp.sort_values(by='SCORE_AV').reset_index()

        ## divide the color to n=100
        pal = sns.color_palette("Reds_d", 100)

        # rank = tmp.score_overall.argsort().argsort()   # http://stackoverflow.com/a/6266510/1628638

        fig = plt.figure(figsize=(8,4))
        ax = fig.add_subplot(111)
        plt.ylim(0,1)

        xticks={}
        left=0
        pos=0

        tmp.loc[:,'color_index']=tmp.apply(lambda row: int(row['SCORE_AV']*100) ,axis=1)

        for count,(group,df) in enumerate( tmp.groupby('SUBJECT_GROUP')):
            
            bars=plt.bar(list(range(0,len(df['SCORE_AV']))), df['SCORE_AV'],color=np.array(pal[::-1])[np.clip(df.color_index, 1, 100)])
            xticks.update({group:left+len(df)//2})
            left=left+len(df)

            ## add tooltips to each bar
            for i, bar in enumerate(bars.patches[0:len(df)]):
                tooltip = mpld3.plugins.LineLabelTooltip(bar, label=df['SUBJECT_NAME'].iloc[i])
                mpld3.plugins.connect(plt.gcf(), tooltip)
            ## move each bar
            for bar_ind,bar in enumerate(bars):
                bar.set_x(pos)
                pos=pos+1



        fontsize=14

        ax.set_xticks(list(xticks.values()))
        ax.set_xticklabels(list(xticks.keys()),fontsize=fontsize)
        plt.setp(plt.xticks()[1], rotation=90)
        plt.ylabel("Gaze quality score!",fontsize=fontsize)
        plt.xlabel("Group",fontsize=fontsize)
        plt.grid(False)

        plt.rcParams['axes.grid'] = False
        plt.xlim(0,len(tmp))
        plt.tight_layout()
        plt.show()




    
```
